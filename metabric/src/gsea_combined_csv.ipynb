```python
import pandas as pd
import numpy as np
```

```python
tp53 = pd.read_csv("GSEA_/TP53/gseapy.gsea.gene_set.report.csv", index_col=0)
brca1 = pd.read_csv("GSEA_/BRCA1/gseapy.gsea.gene_set.report.csv", index_col=0)
brca2 = pd.read_csv("GSEA_/BRCA2/gseapy.gsea.gene_set.report.csv", index_col=0)
rb1 = pd.read_csv("GSEA_/RB1/gseapy.gsea.gene_set.report.csv", index_col=0)
pik3ca = pd.read_csv("GSEA_/PIK3CA/gseapy.gsea.gene_set.report.csv", index_col=0)
gata3 = pd.read_csv("GSEA_/GATA3/gseapy.gsea.gene_set.report.csv", index_col=0)



tp53_fdr = tp53["fdr"].rename("TP53")
brca1_fdr = brca1["fdr"].rename("BRCA1")
brca2_fdr = brca2["fdr"].rename("BRCA2")
rb1_fdr = rb1["fdr"].rename("RB1")
pik3ca_fdr = pik3ca["fdr"].rename("PIK3CA")
gata3_fdr = gata3["fdr"].rename("GATA3")




fdr_df = pd.concat([brca1_fdr, tp53_fdr.reindex(brca1_fdr.index)], axis=1)
fdr_df = pd.concat([fdr_df, brca2_fdr.reindex(fdr_df.index)], axis = 1)
fdr_df = pd.concat([fdr_df, rb1_fdr.reindex(fdr_df.index)], axis = 1)
fdr_df = pd.concat([fdr_df, pik3ca_fdr.reindex(fdr_df.index)], axis = 1)
fdr_df = pd.concat([fdr_df, gata3_fdr.reindex(fdr_df.index)], axis = 1)



fdr_df.sort_values(by="TP53", ascending = True)
```

```python
fdr_df["BRCA1"] = fdr_df.apply(lambda row: 5 if (row["BRCA1"] == 0.0) else (-np.log10(row["BRCA1"])), axis = 1)
fdr_df["BRCA2"] = fdr_df.apply(lambda row: 5 if (row["BRCA2"] == 0.0) else (-np.log10(row["BRCA2"])), axis = 1)
fdr_df["RB1"] = fdr_df.apply(lambda row: 5 if (row["RB1"] == 0.0) else (-np.log10(row["RB1"])), axis = 1)
fdr_df["PIK3CA"] = fdr_df.apply(lambda row: 5 if (row["PIK3CA"] == 0.0) else (-np.log10(row["PIK3CA"])), axis = 1)
fdr_df["GATA3"] = fdr_df.apply(lambda row: 5 if (row["GATA3"] == 0.0) else (-np.log10(row["GATA3"])), axis = 1)
fdr_df["TP53"] = fdr_df.apply(lambda row: 5 if (row["TP53"] == 0.0) else (-np.log10(row["TP53"])), axis = 1)
```

```python
df_index = fdr_df.index.tolist()
new_df_index = []
for i in df_index:
    i = i.rstrip()
    new_df_index.append(i)
fdr_df.index = new_df_index
```
