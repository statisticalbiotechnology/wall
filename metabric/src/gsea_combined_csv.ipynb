```python
import pandas as pd
import numpy as np
```

```python
tp53 = pd.read_csv("GSEA_/TP53/gseapy.gsea.gene_set.report.csv", index_col=0)
brca1 = pd.read_csv("GSEA_/BRCA1/gseapy.gsea.gene_set.report.csv", index_col=0)
tp53_fdr = tp53["fdr"].rename("TP53")
brca1_fdr = brca1["fdr"].rename("BRCA1")
fdr_df = pd.concat([brca1_fdr, tp53_fdr.reindex(brca1_fdr.index)], axis=1)
fdr_df.sort_values(by="TP53", ascending = True)
```

```python
fdr_df.iloc[:50]
```

```python
fdr_df["BRCA1"] = fdr_df.apply(lambda row: 5 if (row["BRCA1"] == 0.0) else (-np.log10(row["BRCA1"])), axis = 1)
fdr_df["TP53"] = fdr_df.apply(lambda row: 5 if (row["TP53"] == 0.0) else (-np.log10(row["TP53"])), axis = 1)
```

```python
fdr_df.dtypes
```

```python
df_index = fdr_df.index.tolist()
new_df_index = []
for i in df_index:
    i = i.rstrip()
    new_df_index.append(i)
fdr_df.index = new_df_index
```
