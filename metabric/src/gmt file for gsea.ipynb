```python
dictionary = {}
for index, row, in reactome.iterrows():
    genes = ast.literal_eval(row['illumina'])
    genelist = []
    for i in genes:
  
            genelist.append(i)
    dictionary[index] = genelist

dictionary
```

```python
with open("../data/GSEA_reactome.gmt", 'w', encoding="utf-8") as f:
    for i in dictionary:
        f.write(f"{i} \t")
        for k in dictionary[i]:
            f.write(k + "\t")
            
        f.write("\n")
    f.close()
```
