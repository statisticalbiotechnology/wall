```python
clusternames = ['1', "2", "3", "4ER+", "4ER-", "5", "6", "7", "8", "9", "10"]

gsea_frames = {}
for cluster in clusternames:
    df = pd.read_csv(f"GSEA_/{cluster}/gseapy.gsea.gene_set.report.csv").iloc[:,:-3]
    #qv.qvalues(df, 'pval', f'cluster {cluster} q-value')
    #df[f'cluster {cluster} q-value'] = inf if x <= 0.0 else -np.log10(x) for x in df[f'cluster {cluster} q-value']
    print(df['fdr'].dtypes)
    df['fdr'] = df.apply(lambda row: 5.0 if (row['fdr'] == 0.0) else -np.log10(row['fdr']), axis=1)
    newlist = []
    for row in df['Term'].iteritems():
        row = row[1].split(" ")[-1:]
        newrow = " ".join(row)
        newlist.append(newrow)

    new_series = pd.Series(newlist)
    df['Term'].update(new_series)
    df = df.set_index('Term')
    df = df.loc[~df.index.duplicated(keep='first')]
    gsea_frames[cluster] = df
```
