```python
cluster_1 = pd.read_csv("GSEA_/1/gseapy.gsea.gene_set.report.csv", index_col = 0)
```

```python

pvalues = cluster_1['fdr'].tolist()
pvalues = [1e-8 if x == 0.0 else x for x in pvalues]##p = 0 gives error so setting it to 1e-320
pvalues = [0.999 if x == 1.0 else x for x in pvalues] ##pathways with p-val 1 gave error as range would be to inf
pvalues = sorted(pvalues)
z_scores = norm.ppf(pvalues)

fig, ax = plt.subplots(1, 1, figsize=(15, 10))
plt.hist(z_scores, bins=50, density=True)
plt.show()
```
